```python
import pandas as pd
```

```python
import plotly.express as px
import matplotlib.pyplot as plt  # Importando Matplotlib para criação de gráficos

```

```python
caminho_arquivo = "DADOS_CRIMINAIS_01_2023_a_02_2024.csv"  # Substitua com o caminho real do seu arquivo
```

```python
dados = pd.read_csv(caminho_arquivo)  # Importando dados do CSV para um DataFrame
```

```python
dados
```

```python
df_valinhos = dados[dados['REGIAO'].str.contains('Valinhos', case=False)]  # Filtrando por "Valinhos" (ignorando maiúsculas/minúsculas)
```

```python
df_valinhos
```

```python
df_campinas = dados[dados['REGIAO'].str.contains('Campinas', case=False)]  # Filtrando por "Valinhos" (ignorando maiúsculas/minúsculas)
df_campinas
```

```python
df_campinas.groupby('NATUREZA2')['NATUREZA2'].count().plot(kind='bar')  # Gráfico de barras por NATUREZA2
plt.title('Contagem por Natureza do Crime em Campinas')  # Título do gráfico
plt.xlabel('Natureza do Crime')  # Rótulo do eixo X
plt.ylabel('Contagem')  # Rótulo do eixo Y
plt.show()  # Exibindo o gráfico
```

```python
dados_agrupados = dados.groupby(["MES", "NATUREZA2"])["MES"].count()
dados_agrupados = dados_agrupados.unstack().fillna(0)  # Transformar em DataFrame multi-índice
```

```python
plt.figure(figsize=(12, 6))  # Definir tamanho do gráfico
dados_agrupados.plot(kind="bar", stacked=True)  # Criar gráfico de barras empilhadas
plt.xlabel("Mês")  # Definir rótulo do eixo X
plt.ylabel("Número de Casos de Crimes")  # Definir rótulo do eixo Y
plt.title("Crimes por Mês e Natureza")  # Definir título do gráfico
plt.xticks(rotation=0)  # Manter rótulos do eixo X na horizontal
plt.legend(title="Natureza", loc="upper left", bbox_to_anchor=(1, 1))  # Adicionar legenda
plt.tight_layout()  # Ajustar layout do gráfico
plt.show()  # Exibir o gráfico
```

```python
# Assuming "Data" is in a format like "DD/MM/YYYY"
dados["Data"] = pd.to_datetime(dados["Data"], format="%d/%m/%Y")  # Convert to datetime
dados["Data"] = dados["Data"].dt.year  # Extract year

dados_agrupados_ano = dados.groupby("Data")["Data"].sum()  # Group by year and sum
plt.figure(figsize=(12, 6))
dados_agrupados_ano.plot(kind="line")
plt.xlabel("Ano")
plt.ylabel("Número Total de Casos")
plt.title("Evolução Anual do Número de Casos")
plt.tight_layout()
plt.show()
```
